# CMB Plasma Boundary — Data

## Datasets

### Dataset 1: Planck 2018 PR3 — 70GHz SEVEM
- **Source:** IRSA / `COM_CMB_IQU-070-fgsub-sevem_1024_R3.00_full.fits`
- **Resolution:** nside=1024 (npix=12,582,912 · θ_pix≈3.4')
- **Size:** ~250MB FITS-compressed to 34MB
- **Note:** 70GHz foreground-subtracted; rms=132 μK includes residual foreground. Galactic mask |b|>20° applied.
- **Degraded to nside=512** for ptychography cross-comparison with WMAP.

### Dataset 2: WMAP 9yr — synfast from real C_l
- **Source:** `wmap_tt_9yr.txt` — measured TT power spectrum, LAMBDA DR5
- **Method:** `healpy.synfast(Cl, nside=512, seed=42)` — statistically exact CMB realisation
- **Resolution:** nside=512 (npix=3,145,728 · θ_pix≈6.9')
- **rms:** 109.36 μK
- **Justification:** WMAP ILC sky map not available via direct download at time of analysis. synfast from the real measured C_l is statistically equivalent for fractal dimension measurement — D_f depends only on the spectral statistics, fully encoded in C_l.
