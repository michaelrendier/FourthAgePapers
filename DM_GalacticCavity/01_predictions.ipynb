## Prediction 2: Flat Velocity Ceiling

The flat rotation velocity is the engine idle RPM:

$$V_{\rm flat} = V_{\rm max,baryonic} \times \Omega_{ZS} = V_{\rm max} \times 0.56714$$

This gives the Tully-Fisher relation from first principles:

$$V_{\rm flat}^4 \propto M_{\rm baryonic} \times \Omega_{ZS}^4$$

The exponent 4 (not 3 or 5) comes from the cavity amplitude ceiling.

```python
# Prediction 2: Tully-Fisher exponent and OMEGA_ZS factor
TF_exponent = 4  # from cavity mode structure
TF_factor   = OMEGA_ZS**4

print(f'Tully-Fisher exponent (predicted):  {TF_exponent}')
print(f'OMEGA_ZS^4 = {TF_factor:.6f}')
print()
# McGaugh (2016) measured: V_flat^4 = G * a_0 * M_baryonic
# where a_0 = 1.2e-10 m/s^2 (MOND acceleration scale)
# SMMIP: a_0 = G * rho_0 * OMEGA_ZS^4 where rho_0 is the mean cosmological density
a0_mond = 1.2e-10  # m/s^2
print(f'MOND a_0 (empirical): {a0_mond:.2e} m/s^2')
print(f'SMMIP prediction: a_0 = (G * H0^2 / (8*pi)) * OMEGA_ZS^4')
H0 = 67.4e3 / 3.086e22  # Planck H0 in s^-1
G  = 6.674e-11
a0_smmip = (3 * H0**2 / (8 * np.pi)) * G * OMEGA_ZS**4 * (3e22)**2  # rough estimate
# This is an order-of-magnitude check only — formal derivation in paper
print(f'SMMIP a_0 (order of magnitude): {a0_smmip:.2e} m/s^2  (rough check)')
```
