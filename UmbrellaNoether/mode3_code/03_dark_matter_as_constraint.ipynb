**Prediction:** Cosmological simulations using Noether-conserving gravitational potentials will reproduce observed galactic rotation curves **without** invoking dark matter particles.

1. **Rotation curves match observations** without dark matter particles
2. **Halo density profiles** match observed NFW profiles

4. **Smooth dynamics** (no singularities in gravitational field)

- Test at multiple galaxy masses and morphologies
