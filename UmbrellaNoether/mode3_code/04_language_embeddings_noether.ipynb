# 04 — Language Embeddings from Noether Balance
## Learning Semantic Structure Without Labels

**Hypothesis:** Word embeddings that enforce Noether balance ($\partial_\mu J^\mu = 0$) will spontaneously develop linguistic structure.

**Prediction:** Embeddings trained under explicit Noether constraints will outperform standard embeddings (Word2Vec, GloVe) on linguistic benchmarks.

## Three Semantic Currents

For each word w:

- **J_R(w)** — Semantic assertion. What the word *means* (primary sense).
- **J_B(w)** — Semantic boundary. What the word *cannot mean* (antonyms, forbidden associations).
- **J_3(w)** — Context sensitivity. How the word's meaning *rotates* with grammar/discourse (polysemy).

**Noether balance:** J_R(w) + J_B(w) + J_3(w) = 0  ⟹  Unique embedding vector

## Expected Results

1. **Word similarity scores** match human judgments (R > 0.85 vs. ~0.70 for GloVe)
2. **Analogy tasks** improve 10–15% over baseline
3. **Antonym relationships** explicitly encoded (opposite J_R values)
4. **Polysemy handled** via J_3 (context-dependent rotation)

## Implementation

The Noether constraint is enforced as a regularization term:

$$L = L_{\text{context}} + \lambda \|J_R + J_B + J_3\|^2$$

where:
- $L_{\text{context}}$ is standard skip-gram or CBOW loss
- $\lambda$ weights the Noether conservation penalty
- The second term forces the three currents to balance
